```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Task 1

```python
x = np.linspace(0,30,100)
rv = sts.chi2(7) # Распределение хи-квадрат с параметром 7
cdf = rv.cdf(x)
plt.plot(x, cdf, label="$k=%s$" % 7)
plt.legend()
plt.title("CDF ($\chi^2_k$)")
```

Сгенерируем выборку

```python
choose = rv.rvs(1000)
plt.hist(choose, normed=True)
plt.ylabel('f(x)')
plt.xlabel('$x$')

# Теоритическая плотность распределения
x = np.linspace(-5,25,100)
pdf = rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
```

n = 5

```python
choose = []
for i in range(1000):
    cur = rv.rvs(5)
    cur = sum(cur)
    choose.append(cur / 5.)
plt.hist(choose, normed=True)
plt.ylabel('f(x)')
plt.xlabel('$x$')

mu = 7 # mu = k
sigma = (14 / 5.) ** 0.5 # mu = 2 * k / n

# зададим нормально распределенную случайную величину
norm_rv = sts.norm(loc=mu, scale=sigma)
x = np.linspace(0, 15, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)
```
